```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x0 = 50
time_programme = 6000
time_pub = 2000
```

```python
list = [x0]
for i in range(time_programme-1):
    list.append(np.random.normal(list[-1],.13))

list.append(list[-1]*0.85)
for i in range(time_pub-1):
    list.append(np.random.normal(list[-1]*0.9999,.12))
list.append(list[-1]*1.15)
for i in range(time_programme-1):
    list.append(np.random.normal(list[-1],.13))    
print(len(list))
```

```python
plt.plot(list)
```
