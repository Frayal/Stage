```python
import xmltodict, json
from xml.dom import minidom 
tree = ET.parse(PATH).getroot()
xmlstr = ET.tostring(tree)
o = xmltodict.parse(xmlstr)
L = json.dumps(o) 
with open('data.txt', 'w') as outfile:
    json.dump(L, outfile)
```

```python
tree = ET.parse(PATH)
root = tree.getroot()
#print(root.tag)
l1 = [child for child in root]
#print(l1)
l2 = [[child for child in root] for root in l1 ]
print(l2[1])

for i,elem in enumerate(l2[1]):
    xmlstr = ET.tostring(elem)
    o = xmltodict.parse(xmlstr)
    with open('data'+str(i)+'.txt', 'w') as outfile:
        json.dump(o, outfile)
    
```
