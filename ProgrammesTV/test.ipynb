```python
#################################################
#created the 25/04/2018 11:10 by Alexis Blanchet#
#################################################
import numpy as np
import pandas as pd
import xml.etree.ElementTree as ET
import xmltodict, json
from collections import OrderedDict
PATH = '/home/alexis/Bureau/Stage/ProgrammesTV/pluri_201712/2017-12-01_programmeTV/IPTV_0192_2017-12-01_TF1.xml'


def load_dic(dic):
    for key in dic["DIFFUSION"]:
        if(len(str(dic["DIFFUSION"][key])) > 50):
            #dic["DIFFUSION"][key] = [(dict(dic["DIFFUSION"][key]))]
            #print(dic["DIFFUSION"][key])
            pass
        else:
            pass
    return dic




tree = ET.parse(PATH)
root = tree.getroot()


#print(root.tag)
l1 = [child for child in root]
#print(l1)
l2 = [[child for child in root] for root in l1 ]
#print(l2[1])

data = pd.DataFrame()
for i,elem in enumerate(l2[1]):
    xmlstr = ET.tostring(elem)
    o = xmltodict.parse(xmlstr)
    o = load_dic(o)
    df2 = pd.DataFrame.from_dict(o["DIFFUSION"],orient = 'index').rename(columns={0: i})
    data = [data,df2.T]
    data = pd.concat(data)
    
    
temp_data = pd.DataFrame()   
for i in range(29):    
    df = data['ATTRIBUTS']
    l = (df[[i]].values)[0]
    df2 = pd.DataFrame.from_dict(l,orient = 'index')
    df2 = df2.rename(columns = {0: i})
    #print(df2.head())
    temp_data = [temp_data,df2]
    temp_data = pd.concat(temp_data,axis = 1)
temp_data.head()
data = data.join(temp_data.T)
temp_data = pd.DataFrame() 
for i in range(29):    
    df = data['EMISSION']
    l = (df[[i]].values)[0]
    df2 = pd.DataFrame.from_dict(l,orient = 'index')
    df2 = df2.rename(columns = {0: i})
    #print(df2.head())
    temp_data = [temp_data,df2]
    temp_data = pd.concat(temp_data,axis = 1)
    temp_data = temp_data.drop(['@DATEMODIF'])

temp_data.head()
data = data.join(temp_data.T)
temp_data = pd.DataFrame() 
data = data.drop(['ATTRIBUTS','EMISSION'],axis = 1)

for i in range(29):    
    df = data['FORMAT']
    l = (df[[i]].values)[0]
    df2 = pd.DataFrame.from_dict(l,orient = 'index')
    df2 = df2.rename(columns = {0: i})
    #print(df2.head())
    temp_data = [temp_data,df2]
    temp_data = pd.concat(temp_data,axis = 1)
temp_data = temp_data.drop(['#text'])
temp_data = temp_data.T.rename(columns={"@CLE": "CLE-FORMAT"})

temp_data.head()
data = data.join(temp_data)
temp_data = pd.DataFrame() 
data = data.drop(['FORMAT'],axis=1)

for i in range(29):    
    df = data['GENRE']
    l = (df[[i]].values)[0]
    df2 = pd.DataFrame.from_dict(l,orient = 'index')
    df2 = df2.rename(columns = {0: i})
    #print(df2.head())
    temp_data = [temp_data,df2]
    temp_data = pd.concat(temp_data,axis = 1)
    temp_data = temp_data.drop(['@GENRESIMPLE'])

temp_data = temp_data.T.rename(columns={"@CLE": "CLE-GENRE","#text":"description programme"})
temp_data.head()
data = data.join(temp_data)
temp_data = pd.DataFrame() 

data = data.drop(['GENRE'],axis=1)
data = data.drop(['LESPERSONNES','LESTEXTES','PHOTO','MAGAZINEXY','SERIE','NATIONALITE','CS','FILMTELEFILM'],axis=1)

data['DATE'] = data['DATEHEURE'].apply(lambda x: x.split('T')[0])
data['HEURE'] = data['DATEHEURE'].apply(lambda x: x.split('T')[1])
data['minutes'] = data['HEURE'].apply(lambda x: int(x.split(':')[0])*60 + int(x.split(':')[1]))
data
    
    
data.to_csv('IPTV_0192_2017-12-01_TF1.csv')
data
```
